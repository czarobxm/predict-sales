## Basic data cleaning process
Methods copied from prep_data.ipynb, where you can read about jutification of below process. I do not include here correlation analisys, as it deletes many attributes important in the process of data analisys.

```python
if not pathlib.Path('data').exists():
    unzip_data("data_zip.zip")
    
data = pd.read_csv('data/project2_training_data.csv')

for col in data.columns:
    print(col)

# Drop nans in address_city and company_id columns
data = data.dropna(subset=['address_city', 'company_id'])

# Delete duplicate columns
cols_with_measure_per_100g = []
cols_with_measure_total = []
for name in data.columns:
    if "_100" in name:
        cols_with_measure_per_100g.append(name.replace("_100", ""))
    if "_total" in name:
        cols_with_measure_total.append(name.replace("_total", ""))

cols_to_delete = list(set(cols_with_measure_per_100g).intersection(set(cols_with_measure_total)))
cols_to_delete = [name + "_total" for name in cols_to_delete]

data = data.drop(cols_to_delete, axis=1)
data.shape

# Handle object columns
data = delete_letters(data, 'size')
data = data.fillna(0)
data = cat_into_num(data)
```
